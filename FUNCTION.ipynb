**1. What is the difference between a function and a method in Python?**

A function is a standalone block of code defined independently, not as part of a class.

It can be called directly by its name from anywhere in the program's scope (global, local, or nested).

Functions operate on data passed to them as arguments and do not inherently have access to or modify the state of any particular object.

Example:
    
    def greet(name):
        return f"Hello, {name}!"

    message = greet("Alice")
    print(message)
Method:
A method is a function that is defined within a class.
It is associated with an object (an instance of a class) and is called on that object.

Methods can access and modify the attributes (data) of the object they belong to, as they implicitly receive a reference to the instance (conventionally named self) as their first argument.

Example:

    class Dog:
        def __init__(self, name):
            self.name = name

        def bark(self):
            return f"{self.name} says Woof!"

    my_dog = Dog("Buddy")
    print(my_dog.bark())

**6. Explain the concept of generators in Python and how they are defined.**

Generators in Python are a special type of iterable that produce values lazily, one at a time, rather than storing all values in memory simultaneously. This makes them highly memory-efficient, especially when dealing with large datasets or potentially infinite sequences.

Generators are defined like regular Python functions, but instead of using the return keyword to return a single value and terminate, they use the yield keyword. When a yield statement is encountered, the function produces a value, pauses its execution, and saves its state. When the generator is iterated over again (e.g., using next() or in a for loop), it resumes execution from where it left off.

Example:

def count_up_to(n):

    i = 0
    while i < n:
        yield i
        i += 1

my_generator = count_up_to(5)

print(next(my_generator))

print(next(my_generator))

print(next(my_generator))  

for num in my_generator:

    print(num)

**10. What is the difference between map(), reduce(), and filter() functions in Python?**

The map(), reduce(), and filter() functions in Python are higher-order functions used for functional programming paradigms, each serving a distinct purpose in data transformation and aggregation.

map():

Purpose: Applies a given function to each item in an iterable and returns an iterator yielding the results.

Output: An iterable (e.g., a map object) with the same number of elements as the input iterable, where each element is the result of applying the function to the corresponding input element.

    numbers = [1, 2, 3, 4]
    doubled = list(map(lambda x: x * 2, numbers))
    
filter():

Purpose: Constructs an iterator from elements of an iterable for which a function returns True.

Output: An iterable (e.g., a filter object) containing only those elements from the input iterable for which the provided function returned a truthy value. The length of the output iterable can be less than or equal to the input

    numbers = [1, 2, 3, 4, 5]
    evens = list(filter(lambda x: x % 2 == 0, numbers))

reduce():

Purpose: Applies a function of two arguments cumulatively to the items of an iterable, from left to right, so as to reduce the iterable to a single value.

Output: A single, accumulated value.
Note: reduce() is not a built-in function in Python 3 and needs to be imported from the functools module (from functools import reduce)

    from functools import reduce
    numbers = [1, 2, 3, 4]
    total_sum = reduce(lambda x, y: x + y, numbers)

**5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.**

```python
class FibonacciIterator:
    def __init__(self, n):
        self.n = n
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration

        fibonacci_number = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return fibonacci_number

num_terms = int(input("Enter the number of Fibonacci terms you want: "))
fibonacci_sequence = FibonacciIterator(num_terms)

for number in fibonacci_sequence:
    print(number,end=" ")
```


**6. Write a generator function in Python that yields the powers of 2 up to a given exponent**

```python
def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

exponent = int(input("Enter the exponent to generate powers of 2: "))
for power in powers_of_two(exponent):
    print(power,end=" ")
```

**7. Implement a generator function that reads a file line by line and yields each line as a string**


```python
def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.rstrip('\n')  # Remove the newline character

file_path = 'C:\Users\ankit\OneDrive\Desktop\assigment Question/FUNCTION ASSIGMENT.pdf.txt'

for line in read_file_line_by_line(file_path):
    print(line)
```

**9. Write a Python program that uses map() to convert a list of temperatures from Celsius to Fahrenheit.**


```python
celsius_temperatures = [0, 20, 30, 40, 100]
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))
print("Temperatures in Fahrenheit:", fahrenheit_temperatures)
```

**10. Create a Python program that uses filter() to remove all the vowels from a given string.**
